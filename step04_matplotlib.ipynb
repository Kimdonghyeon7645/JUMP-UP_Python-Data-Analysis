<p><span style="color:orange; font-size:30px; font-weight:bold;">Matplotlib</span></p>
 
- 데이터를 Chart, Plot으로 시각화해주는 라이브러리이다.
- 2003년 0.1 출시, 17년이된 패키지
- 가장 많이 사용되는 시각화 패키지이다.
- 기본 구성 : 그림(figure), 축(axes)
- % matplotlib inline : jupyter notebook 내에서 output을 보여준다.

## Basic Attributes (기본 속성)

- alpha : 투명도
- logy : Y축 Log scaling
- kind : line, bar, barh, kde
- subplots : 여러개의 plot 그리기
- legend : subplot의 범례 지정
- xlim, ylim : X축과 Y축의 경계(axis로 한번에 그릴 수 있음)
- grid : 그리드(격자, 눈금) 표현(True, False)
- title : 그래프의 제목 지정
- linewidth : 라인 넓이
- color : 색
- linestyle : 실선, 점선, 1점 쇄선 등
- marker : 마커 지정
- markerfacecolor : 마커 색
- markersize : 마커 크기
- 그외에도 다양한 속성이 존재 : [튜토리얼 사이트로](https://matplotlib.org/3.1.1/tutorials/introductory/pyplot.html)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[100,200,300]
y=[1,2,3]
plt.plot(x,y)
```

```python
x=[100,200,300]
y=[1,2,3]
print(pd.Series(x, y))
plt.plot(pd.Series(x, y))
```

```python
x=[100,200,300]
y=[1,2,3]
plt.plot(x,y,'-.r')  
# g(green), r(red), b(blue))
# 색깔(g, r, b) 앞에 :을 붙이면 점선으로 라인이 그어짐
# --도 되고, -.도 되고, .도 되고, --> 도 되고, --<도 되고, --o도 되고 이런것을 색깔 앞에 붙여보자
```

```python
x=[100,200,300]
y=[1,2,3]
plt.plot(x,y,color='#ff00ff', linewidth='20')  # #ff00ff(핑크색) 같이 rgb 색상 코드도 사용가능
```

```python
x=[100,200,300]
y=[1,2,3]
plt.plot(x,y,'--r', linewidth='3')
plt.xlabel('X label',fontsize=16)    # 가로 라벨
plt.ylabel('Y Label',fontsize=16)    # 세로 라벨
plt.title('Matplotlib Test',fontsize='20')    # 표 제목
```

```python
x=[100,200,300]
y=[1,2,3]
plt.plot(x,y,'--r', linewidth='3')
plt.xlabel('X label',fontsize=16)    # fontsize를 라벨, 제목에 지정 가능
plt.ylabel('Y Label',fontsize=16)
plt.title('Matplotlib Test',fontsize='20')
plt.savefig('sample.png')    # 같은 경로에서 sample.png 란 이미지로 저장!
```

```python
x=np.linspace(0,10,100)    # 0부터 10까지 수를 100개 만들어서 반환함
siny=np.sin(x)
cosy=np.cos(x)

plt.plot(x,siny, 'b')
plt.plot(x,cosy, 'g')
```
