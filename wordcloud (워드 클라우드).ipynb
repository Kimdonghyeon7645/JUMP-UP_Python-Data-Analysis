```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = '''python python python python python
big data big dat big data big data big data big data big data big data big data big data
crawling crawling crawling crawling
analysis analysis analysis analysis
visualization visualization visualization visualization
machine learning machine learning machine learning machine learning machine learning
deep learning deep learning deep learning deep learning deep learning deep learning
'''

wc = WordCloud(width=1000, height=600, background_color="white", random_state=0, font_path=r'c:\Windows\Fonts\malgun.ttf')
plt.imshow(wc.generate(text))
plt.axis("off")    # 테두리(눈금) 없에기
plt.show()

# 아래와 같이 빈도를 출력
WordCloud().generate(text).words_
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

frequency_d = {
    '동동':15, 
    '취권':1,
    '동동주':4,
    '김동':4,
    '동주':2,
    '동대문':5,
    '동현':6
}

wc = WordCloud(width=400, height=500, background_color="white", random_state=0, font_path=r'c:\Windows\Fonts\malgun.ttf')
plt.imshow(wc.generate_from_frequencies(frequency_d))
plt.axis("off")
plt.show()
```

```python
import numpy as np
from PIL import Image
from wordcloud import STOPWORDS

text=open("worldcloud files/alice.txt").read()

alice_mask=np.array(Image.open("worldcloud files/alice.png"))
stopwords=set(STOPWORDS)
stopwords.add("said")

alice_mask
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
wc=WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords)
wc=wc.generate(text)

wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import random

def gray_color_func(word,font_size,position, orientation, random_state=None,**kwargs):
    return "hsl(0,0%%,%d%%)" % random.randint(60,100)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=gray_color_func, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
text=open("worldcloud files/alice.txt").read()

alice_coloring=np.array(Image.open("worldcloud files/alice_color.png"))
stopwords=set(STOPWORDS)    # 추가하기 싫은 문자(불용어) 생성
stopwords.add("said")       # 불용어에 said 를 추가
```

```python
wc=WordCloud(background_color='white', max_words=2000, mask=alice_coloring, 
             stopwords=stopwords, max_font_size=40,random_state=42)
wc=wc.generate(text)
```

```python
from wordcloud import ImageColorGenerator
image_colors=ImageColorGenerator(alice_coloring)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.imshow(alice_coloring, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis('off')
plt.show()
```
